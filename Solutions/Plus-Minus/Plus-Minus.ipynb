```python
from itertools import groupby

def plusMinus(arr):
    # Write your code here
    base_len = len(arr)
    arr.sort()
    def sorter(x):
        if x > 0: return 'pos'
        elif x < 0: return 'neg'
        else: return 'zero'
    lx = {key: list(val)  for key, val in groupby(arr, key=sorter)}
    div_lx = lambda x: len(lx[x]) / base_len
    for i in ['pos', 'neg', 'zero']: 
        print(div_lx(i)) if i in lx.keys() else print(0.000000)
```

```python
# Create test cases
from random import randrange

test_case = lambda x: [randrange(-100,100) for i in range(randrange(3,x))]
test_array = [test_case(250) for i in range(100)]
for i in test_array: plusMinus(i)
```

```python
from itertools import groupby

def hide_and_seek(s):
    out = [(len(list(g)),int(k)) for k, g in groupby(s, key=lambda x:x[0])]            
    return out


# set up test case data
from random import randrange
test_case = lambda: ''.join([str(randrange(0,10)) for x in range(randrange(10, 20))])
test_array = [test_case() for i in range(25)]

for input_string in test_array: 
    print(input_string)
    print(*hide_and_seek(input_string), sep=' ')
```
